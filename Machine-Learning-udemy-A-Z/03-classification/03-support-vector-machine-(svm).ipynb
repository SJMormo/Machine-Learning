## SVM

## Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import Data Set

```python
dataset = pd.read_csv('./Social_Network_Ads.csv')
X = dataset.iloc[:, 2:4].values # taking age feature
y = dataset.iloc[:, -1].values # taking only the last col - purchase
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

## Feature Scaling
### y is categorical dependent variable so no feature scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Fitting the SVM classifier to the training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=0) # Try out different kernel to improve the model
classifier.fit(X_train, y_train)
```

## Predict

```python
y_pred = classifier.predict(X_test)
```

## Confusion Matrix
This matrix contains the correct predictions that our model made as well the incorrect ones to evaluate predictive power of out model

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Visualize the training set

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min()-1, stop=X_set[:, 0].max()+1, step=0.01),
    np.arange(start=X_set[:, 1].min()-1, stop=X_set[:, 1].max()+1, step=0.01)
)
plt.contourf(
    X1, X2, classifier.predict(
        np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), 
        alpha = 0.75, cmap = ListedColormap(('red', 'green')
    )
)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
    c = ListedColormap(('red', 'green'))(i), label = j)

plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualize the test result

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min()-1, stop=X_set[:, 0].max()+1, step=0.01),
    np.arange(start=X_set[:, 1].min()-1, stop=X_set[:, 1].max()+1, step=0.01)
)
plt.contourf(
    X1, X2, classifier.predict(
        np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), 
        alpha = 0.75, cmap = ListedColormap(('red', 'green')
    )
)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
    c = ListedColormap(('red', 'green'))(i), label = j)

plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
